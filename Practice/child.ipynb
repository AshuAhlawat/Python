# Nutrients Calculator

### This functions is used to find out if your daily calorie intake is healthy according to your age by selecting **(1)** in the main menu.

```python
def nutri_calc():
    # by age classification of nutrients required for healthy growth
    required_nutrients = {
        "0-2": 800,
        "2-4": 1400,
        "4-8": 1800
    }
            
    if age <2:
        target = required_nutrients["0-2"]
    elif age <4:
        target = required_nutrients["2-4"]
    else:
        target = required_nutrients["4-8"]

    # foods with calorie per gram
    foods = [
        ("Milk",1.0),
        ("Egg", 1.55),
        ("Rice", 1.3),
        ("Lentils", 1.13), 
        ("Vegetables", 0.85), 
        ("Meat",1.43)
    ]

    calories = 0 

    print("Enter child's daily consumption of :")

    for food,cal_per_gm in foods:
        while True:
            try:
                diet = int(input(food+"(in grams) : "))
                calories += diet*cal_per_gm
                break
            except:
                print("Wrong Input! Enter again ", end="")


    print("Today you consumed "+ str(int(calories)) + " calories")
    if calories >= target:
        print("You have a healthy diet for a " + str(age)+ " year old child!")
    else:
        print("You are unfortunately undernourished and should eat atleast " + str(int(target-calories)) + " more calories per day")
```

# Basic Info

### this function saves your name and age, which can be re-written by chosing **(3)** in the main menu

```python
def info():
    while True:
        try:
            name = str(input("What should i call you? -"))
            if name:
                break
            else:
                print("Please enter your name")
        except:
            print("Wrong Input, Try again")
    
    while True:
        try:
            age = int(input("How old are you? -"))
            
            if age <= 8:
                break
            else:
                print("You are not a child.")
        except:
            print("Wrong Input, Try again")
    
    print("Hi " + name + ", what would you like to calculate?")
    return name,age
```
